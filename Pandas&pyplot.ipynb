```python
sns.set(style="whitegrid")
plt.figure(figsize=(20, 16))
sns.countplot(x="Discount_offered", hue="Gender", data=car)
plt.title('Count of Discount Offered by Gender')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(30, 20))
ax = sns.countplot(x="Discount_offered", hue="Gender", data=car)

# Annotate each bar with the actual count
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Count of Discount Offered by Gender')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.countplot(x="Customer_care_calls", hue="Gender", data=car)

# Annotate each bar with the actual count
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Count call by gender by Gender')
plt.show()
```
